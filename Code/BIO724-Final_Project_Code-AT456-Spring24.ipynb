```python
def getHeadMean(day_index, dataframe, condition): #calculates mean of the 28 day 'head'
    return(dataframe[condition][day_index:day_index+28].mean())

def getHeadSD(day_index, dataframe, condition): #calculates sd of the 28 day 'head'
    return(dataframe[condition][day_index:day_index+28].std())

def getTailMean(day_index, dataframe, condition): #calculates mean of the 3 day 'tail'
    return(dataframe[condition][day_index+28:day_index+31].mean())

def whiplash(day_index, dataframe, condition): #calculates (MeanHead - MeanTail) / SDHead
    return((getHeadMean(day_index, dataframe, condition)-getTailMean(day_index, dataframe, condition))/getHeadSD(day_index, dataframe, condition))
```

```python
# check the functions
print(getHeadMean(0, Redding_2020_2023_python, 'T_DAILY_AVG'))
print(getHeadSD(0, Redding_2020_2023_python, 'T_DAILY_AVG'))
print(getTailMean(0, Redding_2020_2023_python, 'T_DAILY_AVG'))
print(whiplash(0, Redding_2020_2023_python, 'T_DAILY_AVG'))
```

```python
# Makes a function that will call the whiplash function up until the length of the dataframe.
# All data will be appended to list 'y' consecutively.
# Creates a new column titled 'condition_whiplash' which contains NaNs in all cells.
# Data from list y is added to column 'condition_whiplash' of the dataframe starting at the 28th row (29th day), 
    # which is where the Tail, or 'whiplash window' starts

def getListofwhiplash(startVal, dataframe, condition):
    x = dataframe
    y = []
    counter = startVal
    while counter <= len(x)-29:
        y.append(whiplash(counter, x, condition))
        counter = counter + 1
    x[condition+'_whiplash'] = np.nan
    x.iloc[28:len(y)+28, -1] = y
    return(x)
```
